## Import Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Replace this with your actual path and filename
file_path = r'C:\Users\nolia\OneDrive\Desktop\spotify-data-visualization\spotify-2023.csv'

# Load the dataset
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Preview the data
df.head()
```

```python
df.head()
df.info()
df.describe(include='all')
```

## Data Cleaning

```python
### clean 'streams' column 
df['streams_clean'] = df['streams'].str.replace(',', '', regex=False)
df = df[df['streams_clean'].str.isnumeric()].copy()
df['streams'] = df['streams_clean'].astype('int64')
df.drop(columns='streams_clean', inplace=True)
```

```python
### check for duplicates
print(f'Duplicates before drop: {df.duplicated().sum()}')

# drop if any
df = df.drop_duplicates()
```

```python
# Check nulls
nulls = df.isnull().sum()
print(nulls[nulls > 0])
```

```python
### clean shazam column
# convert to string and remove commas
df['in_shazam_charts'] = df['in_shazam_charts'].astype(str).str.replace(',', '', regex=False)

# replace 'nan' string with '0'
df['in_shazam_charts'] = df['in_shazam_charts'].replace('nan', '0')

# convert to integer
df['in_shazam_charts'] = df['in_shazam_charts'].astype('int64')
```

```python
### clean key column by dropping nans, since they are an important musical property and should not be guessed
df = df.dropna(subset=['key'])
```

```python
df.isnull().sum()
```

```python
### check all numeric columns are actually numeric
df.dtypes
```

```python
### clean deezer column
# convert to string and remove commas
df['in_deezer_playlists'] = df['in_deezer_playlists'].astype(str).str.replace(',', '', regex=False)

# replace 'nan' string with '0'
df['in_deezer_playlists'] = df['in_deezer_playlists'].replace('nan', '0')

# convert to integer
df['in_deezer_playlists'] = df['in_deezer_playlists'].astype('int64')
```

```python
### check for unique values
for col in ['key', 'mode', 'released_year', 'artist_count']:
    print(f'\n{col}:\n', df[col].unique())
```

## Visualizations

```python
# Set style
sns.set(style='whitegrid', palette='muted')

# Features to analyze
features = ['danceability_%', 'energy_%', 'valence_%', 'acousticness_%']

# Plot each feature against streams
for feature in features:
    plt.figure(figsize=(8, 5))
    sns.regplot(data=df, x=feature, y='streams', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
    plt.title(f'Streams vs {feature}', fontsize=14)
    plt.xlabel(feature.replace('_', ' ').title())
    plt.ylabel('Streams')
    plt.tight_layout()
    plt.show()
```

```python
# list of platforms to analyze
placement_cols = ['in_spotify_playlists', 'in_apple_playlists', 'in_shazam_charts']

# loop through each platform and plot average streams
for col in placement_cols:
    plt.figure(figsize=(6, 4))
    
    # group and average
    avg_streams = df.groupby(df[col] > 0)['streams'].mean().reset_index()
    avg_streams[col] = avg_streams[col].map({True: f'{col} > 0', False: f'{col} = 0'})
    
    # bar plot
    sns.barplot(data=avg_streams, x=col, y='streams')
    plt.title(f'Average Streams vs {col.replace("_", " ").title()} Presence')
    plt.ylabel('Average Streams')
    plt.xlabel('Playlist/Chart Presence')
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# list of playlist/chart columns
placement_cols = ['in_spotify_playlists', 'in_apple_playlists', 'in_shazam_charts']

for col in placement_cols:
    plt.figure(figsize=(12, 5))

    # plot count of songs
    plt.subplot(1, 2, 1)
    counts = (df[col] > 0).value_counts()
    counts.index = counts.index.map(lambda x: f'{col} > 0' if x else f'{col} = 0')
    sns.barplot(x=counts.index, y=counts.values)
    plt.title(f'Number of Songs by {col.replace("_", " ").title()} Presence')
    plt.ylabel('Count')
    plt.xlabel('')

    # plot stream distribution
    plt.subplot(1, 2, 2)
    df['placement_group'] = df[col].apply(lambda x: f'{col} > 0' if x > 0 else f'{col} = 0')
    sns.boxplot(data=df, x='placement_group', y='streams')
    plt.title(f'Streams Distribution by {col.replace("_", " ").title()}')
    plt.ylabel('Streams')
    plt.xlabel('')
    
    plt.tight_layout()
    plt.show()
```

```python
# create a new column counting how many platforms each song appears on
playlist_cols = ['in_spotify_playlists', 'in_apple_playlists', 'in_shazam_charts']
df['playlist_count'] = df[playlist_cols].gt(0).sum(axis=1)

# visualize boxplot of streams by number of playlists
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='playlist_count', y='streams')
plt.title('Streams vs Number of Playlist/Chart Features')
plt.xlabel('Number of Platforms (Spotify, Apple, Shazam)')
plt.ylabel('Streams')
plt.tight_layout()
plt.show()
```
